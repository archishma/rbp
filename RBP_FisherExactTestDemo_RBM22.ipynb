```python
# following Charlene's tutorial 
# the goal is to find genes essential to HepG2 (cancer cell line)

import pandas as pd
from pybedtools import BedTool
dep_score = pd.read_csv('D2_DRIVE_gene_dep_scores.csv')
dep_score
```

```python
# split the names out
dep_score['Symbol'] = dep_score['Unnamed: 0'].str.split(' ', expand = True)[0]
```

```python
# drop the unneeded column and make the index Symbol (labels)
dep_score.drop('Unnamed: 0', inplace = True, axis = 1)
dep_score.set_index('Symbol', inplace = True)
dep_score
```

```python
# get hepg2 targets by making a hepg2 df
hepg2 = dep_score[['HEPG2_LIVER']]
hepg2
```

```python
targets = hepg2.loc[hepg2['HEPG2_LIVER'] < -1].index
type(targets)
# index is a sequence used for sequencing, immutable like string
```

```python
targets
```

```python
# splitting out the & from targets 
essential_gene = []
for t in targets:
    if '&' in t:
        essential_gene.extend(t.split('&'))
        print(t)
    else:
        essential_gene.append(t)
```

```python
essential_gene
# these are the transcripts bound
# in ENCODE, we have RBPs 
```

## Goal:
- does our RBP (RBM22) regulate tumor-essential genes more than other genes? (likelihood)

4 groups we need to find:
1. tumor essential AND bound by RBM22
2. tumor essential and NOT bound by RBM22
3. NOT tumor essential AND bound by RBM22
4. NOT tumor essential and NOT bound by RBM22

Use BedTool intersect to find the 4 groups:

```python
# group 1: essential genes AND bound by RBM22
group_1 = essential_gene_coords.intersect(idr, s = True, u = True).saveas()
```

```python
# group 2: essential genes and NOT bound by RBM22
group_2 = essential_gene_coords.intersect(idr, s = True, v = True).saveas()
```

```python
# transcripts that are not tumor essential genes, so we can repeat this process for group 3 & 4
nonessential_gene_coords = transcript_coords.intersect(essential_gene_coords, s = True, v = True).saveas()
```

```python
# group 3: nonessential genes AND bound by RBM22
group_3 = nonessential_gene_coords.intersect(idr, s = True, u = True).saveas()
```

```python
# group 4: nonessential genes and NOT bound by RBM22
group_4 = nonessential_gene_coords.intersect(idr, s = True, v = True).saveas()
```

```python
# counting genes for table
table_for_stat = pd.DataFrame([
    [len(group_1), len(group_2)], 
     [len(group_3), len(group_4)]
], 
    
    columns = ['bound by RBM22', 'not bound'],
    index = ['essential', 'not essential']
)
```

```python
table_for_stat
# things to note - syntax for putting in rows [square brackets]
```

Notes from Charlene on stats
- scipy stats documentation: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.fisher_exact.html
- odds ratio from NIH: https://www.ncbi.nlm.nih.gov/pmc/articles/PMC2938757/


```python
from scipy.stats import fisher_exact
```

```python
odds, p = fisher_exact(table_for_stat, alternative = 'two-sided')
```
